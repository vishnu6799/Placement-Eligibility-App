# DATASET CREATION


```python
#!pip install Faker
from faker import Faker
F = Faker()
```

```python
#STUDENT TABLE
'''
student_id (Primary Key): Unique identifier for each student.
● name: Full name of the student.
● age: Age of the student.
● gender: Gender of the student (e.g., Male, Female, Other).
● email: Email address of the student.
● phone: Contact number of the student.
● enrollment_year: Year when the student enrolled in the course.
● course_batch: Name of the batch or cohort the student belongs to.
● graduation_year: Expected or actual graduation year for the student.
● CGPA: CGPA out of 10.
● Attandance: Attandance in %.
'''
import pandas as pd

data =[]
col = ["Student_ID","Name","Age","Gender","Email","Phone","Enrollment_year","Course_Batch","Graduation_Year", "CGPA","Attandance"]
import random
for i in range(1,5120):
    Student_ID = i
    Name = F.name()
    Age = random.randint(22,35)
    Gender = random.choice(['Male','Female'])
    Email = F.email()
    Phone= F.phone_number()
    Enrollment_year = random.randint(2023,2025)
    Course_Batch = random.choice(['Weekday','Weekend'])
    Graduation_Year = random.randint(2018,2025)
    CGPA = round(random.uniform(4.5,10.0),2)
    Attandance = random.randint(60,100)
    data.append([Student_ID,Name , Age, Gender,Email,Phone,Enrollment_year,Course_Batch,Graduation_Year, CGPA,Attandance])
Students_Table = pd.DataFrame(data, columns = col)

Students_Table.head(25)
```

```python
#PROGRAMMING TABLE
'''This table stores details of students' programming performance in the course.
● programming_id (Primary Key): Unique identifier for each programming
record.
● student_id (Foreign Key): References the student_id in the Students
Table.
● language: Programming language being evaluated (e.g., Python, SQL).
● problems_solved: Total number of coding problems solved by the student.
● assessments_completed: Number of assessments completed by the student.
● mini_projects: Number of mini projects submitted by the student.
● quiz score: Average Marks obtained in all the quizzes.'''

col1 = ["Programming_ID","Student_ID","Language","Problems_solved","Assessments_completed",
       "Mini_Projects", "Quiz_Score"]
data = []
for i in range(1,5120):
    Programming_ID = F.uuid4()
    Student_ID = i
    Language = random.choice(['Python','SQL','Java','C++'])
    Problems_solved = random.randint(180, 500)
    Assessments_completed = random.randint(1,10)
    Mini_Projects = random.randint(1,5)
    Quiz_Score = round(random.uniform(5.5,10.0),2)
    data.append([Programming_ID,Student_ID,Language,Problems_solved,Assessments_completed,Mini_Projects, Quiz_Score])

Programming_Table =pd.DataFrame(data, columns = col1) 
    
Programming_Table.head(25)
```

```python
#Soft Skills Table
'''This table stores data on students' performance in soft skills evaluations.
● soft_skill_id (Primary Key): Unique identifier for each soft skill record.
● student_id (Foreign Key): References the student_id in the Students
Table.
● communication: Communication skills score (out of 100).
● teamwork: Teamwork skills score (out of 100).
● presentation: Presentation skills score (out of 100).
● leadership: Leadership skills score (out of 100).
● critical_thinking: Critical thinking skills score (out of 100).
● interpersonal_skills: Interpersonal skills score (out of 100).'''

col2 = ["Soft_Skill_ID","Student_ID","Communication","Team_Work","Presentation","Leadership",
        "Critical_Thinking","Interpersonal_Skills"]
data=[]
for i in range(1,5120):
    Soft_Skill_ID = f"SS_{i:04d}"
    Student_ID = i
    Communication = random.randint(70,100)
    Team_Work = random.randint(50,100)
    Presentation = random.randint(60,100)
    Leadership = random.randint(70,100) 
    Critical_Thinking = random.randint(60,100)
    Interpersonal_Skills = random.randint(60,100)
    data.append([Soft_Skill_ID,Student_ID,Communication,Team_Work,Presentation,Leadership,
        Critical_Thinking,Interpersonal_Skills])

Soft_Skills_Table = pd.DataFrame(data, columns=col2)
Soft_Skills_Table.head(25)
```

```python
#Placements Table
'''This table stores details related to students’ placement readiness and outcomes.
● placement_id (Primary Key): Unique identifier for each placement record.
● student_id (Foreign Key): References the student_id in the Students
Table.
● mock_interview_score: Score in the mock interviews (out of 100).
● internships_completed: Total number of internships completed by the student.
● placement_status: Placement status (e.g., Unplaced, Placed).
● company_name: Name of the company where the student got placed (if
applicable)
● placement_package: Package offered during placement (in USD or local
currency).
● interview_rounds_cleared: Number of interview rounds cleared by the
student.
● placement_date: Date when the placement offer was receive'''

col3 = ["Placement_ID", "Student_ID", "Mock_Interview_Score", "Internships_Completed",
        "Placement_Status", "Company_Name", "Placement_Package_INR_LPA", 
        "Inteview_rounds_cleared", "Placement_date"]

data = []

for i in range(1, 5120):
    Placement_ID = f"PP_{i:04d}"
    Student_ID = i
    Mock_Interview_Score = random.randint(60, 100)
    Internships_Completed = random.randint(0, 2)
    Placement_Status = random.choice(['Placed', 'Unplaced'])
    
    if Placement_Status == 'Placed':
        Company_Name = F.company()
        Placement_Package_INR_LPA = round(random.uniform(3.5, 18.0), 2)
        Placement_Date = F.date_between(start_date='-1y', end_date='today')  

    else:
        Company_Name = "Not Applicable"
        Placement_Package_INR_LPA = "Not Applicable"
        Placement_Date = "Not Applicable"
        
    Interview_Rounds_Cleared = random.randint(0, 5)
    
    data.append([Placement_ID, Student_ID, Mock_Interview_Score, Internships_Completed,
                 Placement_Status, Company_Name, Placement_Package_INR_LPA,
                 Interview_Rounds_Cleared, Placement_Date])

Placements_Table = pd.DataFrame(data, columns=col3)

Placements_Table.head(25)
```

# CONNECTING WITH SQL


```python
import sqlite3
class CampusPlacementsdb:
    def __init__(self,db_name = 'Campus_Placements.db'):
        self.conn = sqlite3.connect(db_name)
        self.cursor = self.conn.cursor()
    
    def create_table(self):
        self.cursor.execute('''
        CREATE TABLE IF NOT EXISTS STUDENTS_TABLE(
        Student_ID INTEGER PRIMARY KEY,
        Name TEXT,
        Age INTEGER,
        Gender TEXT,
        Email TEXT,
        Phone TEXT,
        Enrollment_year INTEGER,
        Course_Batch INTEGER,
        Graduation_Year INTEGER,
        CGPA FLOAT,
        Attandance INTEGER )
        ''')
        
        self.cursor.execute('''
        CREATE TABLE IF NOT EXISTS PROGRAMMING_TABLE(
        Programming_ID TEXT PRIMARY KEY,
        Student_ID INTEGER,
        Language TEXT,
        Problems_solved INTEGER,
        Assessments_completed INTEGER,
        Mini_Projects INTEGER,
        Quiz_Score FLOAT,
        FOREIGN KEY (Student_ID) REFERENCES STUDENTS_TABLE(Student_ID))
        ''')
        
        self.cursor.execute('''
        CREATE TABLE IF NOT EXISTS SOFT_SKILLS_TABLE(
        Soft_Skill_ID TEXT PRIMARY KEY,
        Student_ID INTEGER,
        Communication INTEGER,
        Team_Work INTEGER,
        Presentation INTEGER,
        Leadership INTEGER,
        Critical_Thinking INTEGER,
        Interpersonal_Skills INTEGER,
        FOREIGN KEY (Student_ID) REFERENCES STUDENTS_TABLE(Student_ID))
        ''')
        
        self.cursor.execute('''
        CREATE TABLE IF NOT EXISTS PLACEMENTS_TABLE(
        Placement_ID TEXT PRIMARY KEY,
        Student_ID INTEGER,
        Mock_Interview_Score INTEGER,
        Internships_Completed INTEGER,
        Placement_Status TEXT,
        Company_Name TEXT,
        Placement_Package_INR_LPA TEXT,
        Inteview_rounds_cleared TEXT,
        Placement_date DATE,
        FOREIGN KEY (Student_ID) REFERENCES STUDENTS_TABLE(Student_ID))
        ''')
        
        self.conn.commit()
    
    def insert_data(self,df,table_name):
        df.to_sql(table_name, self.conn, if_exists='replace',index=False)
        print(f"Data inserted into table: {table_name}")
    
    def close(self):
        self.conn.close()
        
```

```python
data_base = CampusPlacementsdb()
data_base.create_table()
```

```python
data_base.insert_data(Students_Table, 'STUDENTS_TABLE')
data_base.insert_data(Programming_Table, 'PROGRAMMING_TABLE')
data_base.insert_data(Soft_Skills_Table, 'SOFT_SKILLS_TABLE')
data_base.insert_data(Placements_Table, 'PLACEMENTS_TABLE')

data_base.close()
```
